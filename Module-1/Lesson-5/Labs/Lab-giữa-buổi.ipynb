```python
import numpy as np
np.random.seed(42)
```

# Problem 1: Tạo hai ma trận 3x3 với các phần tử integer ngẫu nhiên từ 1 - 10.

```python
mat1 = np.random.randint(1, 10, (3, 3))
mat1
```

```python
mat2 = np.random.randint(1, 10, (3, 3))
mat2
```

# Problem 2: Thực hiện phép cộng phần tử, nhân từng phần tử, tích vô hướng với hai ma trận vừa được tạo.


```python
sum_mat = np.add(mat1, mat2)
sum_mat
```

```python
mul_mat = np.multiply(mat1, mat2)
mul_mat
```

```python
matmul_mat = np.matmul(mat1, mat2)
matmul_mat
```

# Problem 3: VIết hàm nhận vào một mảng Numpy, sau đó trả về tổng tất cả các phần tử của chúng


```python
def sum_element(mat): 
    sum = np.sum(mat)
    return sum 

sum_element(mat1)
```
